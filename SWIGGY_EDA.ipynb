```python



top_cuisine_percentages = (cuisines_df[cuisines_df['Cuisines'].isin(top_three_cuisines.index)]
                           .groupby('Cuisines')['Restaurant ID']
                           .nunique()
                           .divide(9551)
                           .multiply(100))
```

```python
# # LEVEL-2-TASK-3
# Task: Geographic Analysis
# Plot the locations of restaurants on a
# map using longitude and latitude
# coordinates.
# Identify any patterns or clusters of
# restaurants in specific areas.
```

```python
# Plot the locations of restaurants on a
# # map using longitude and latitude
# # coordinates
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5, marker='*',c='r')
plt.title('Restaurant Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
# Identify any patterns or clusters of
# # restaurants in specific areas.
```

```python
from sklearn.cluster import KMeans
```

```python
X = df[['Longitude', 'Latitude']]


n_clusters = 5 


kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(10, 6))
for cluster in range(n_clusters):
    cluster_df = df[df['Cluster'] == cluster]
    plt.scatter(cluster_df['Longitude'], cluster_df['Latitude'], label=f'Cluster {cluster}', alpha=0.5, marker='*')
plt.title('Clusters')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Analyze the ratings and popularity of
# different restaurant chains.
```

```python
restaur_stats = df.groupby('Restaurant Name').agg({'Aggregate rating': 'mean', 'Votes': 'sum'})


popular_chains = restaur_stats.sort_values(by=['Votes', 'Aggregate rating'], ascending=False)
```
